# One dimensional pressure distribution in linear flow

Non dimensional pressures are calculated for constant upstream pressure of one.

## Non-dimensional pressure at constant pressure conditions

```python
p1 = OnePhase.solve_ndimpressure_pconst(xvalues,0.01,0)
p2 = OnePhase.solve_ndimpressure_pconst(xvalues,0.05,0)
p3 = OnePhase.solve_ndimpressure_pconst(xvalues,0.1,0)
p4 = OnePhase.solve_ndimpressure_pconst(xvalues,0.2,0)
p5 = OnePhase.solve_ndimpressure_pconst(xvalues,0.5,0)
```

```python
plt.plot(xvalues,p1,label='tD = 0.01')
plt.plot(xvalues,p2,label='tD = 0.05')
plt.plot(xvalues,p3,label='tD = 0.1')
plt.plot(xvalues,p4,label='tD = 0.2')
plt.plot(xvalues,p5,label='tD = 0.5')

plt.xlim((0,1))
plt.ylim((0,1))

plt.legend()

plt.xlabel("xD")
plt.ylabel("PD")

plt.show()
```
